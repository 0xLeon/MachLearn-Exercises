```python
def lm_simple(x,y):
    b1 = (x-x.mean()).dot((Y - Y.mean())) / (x-x.mean()).dot((x - x.mean()))
    b0 = Y.mean() - b1 * x.mean()

    y_hat = b0 + b1 * x
    RSS = (y-y_hat).dot(y-y_hat)
    SE = RSS / (y_hat-y_hat.mean()).dot(y_hat-y_hat.mean()) * 1/(len(x)-2)
    t_value = b1 / SE
    p_value = scipy.stats.norm.pdf(t_value, loc=0, scale=1) 
    return (b1, p_value)
```
